Our next mission is to use the DataFrames containing campaign committees and contributors to figure out the biggest donors both for and against the measure (legalização da maconha na Califórnia).

To do that, the first thing we need to do is isolate the fundraising committees active on Proposition 64, which are now buried among of the list of more than 100 groups active last November.


```python
props[props.prop_name == 'PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE.']
```

```python
prop = props[props.prop_name == 'PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE.']
```

```python
prop.head()
```

```python
prop.info()
```

Our next job is to filter down the contributions list, which includes all disclosed contributions to all proposition campaigns, to just those linked to Proposition 64.


```python
contribs.info()
```

You will notice that this file contains a field called calaccess_committee_id that is identical to the one found in the committee CSV.

That’s because these two files are drawn from a “relational database” that stores data in an array of tables linked together by common identifiers. In this case, the unique identifying codes of committees in one table can be expected to match those found in another.

We can therefore safely join the two files using the pandas merge method.

```python
merged = pd.merge(prop, contribs, on="calaccess_committee_id")
```

```python
merged.head()
```

```python
merged.info()
```

```python
#  What is the total sum of Proposition 64 contributions that have been reported?
    # Apenas a coluna "amount"
merged.amount
```

```python
# Soma
merged.amount.sum()
```

We want to separate the money spent supporting the proposition from the money opposing it. Then we want to find out who raised more.

First let’s look at the column we’re going to filter by, committee_position

```python
merged.committee_position.value_counts()
```

```python
support = merged[merged.committee_position == 'SUPPORT']
```

```python
support.amount.sum()
```

```python
oppose = merged[merged.committee_position == 'OPPOSE']
```

```python
oppose.amount.sum()
```

The support is clearly larger. But what percent is it of the overall disclosed total? We can find out by combining two sum calculations using the division operator.

```python
support.amount.sum() / merged.amount.sum()
```

What were the ten biggest contributions? We can find the answer by using the sort_values method to rearrange our list using the amount field.

```python
merged.sort_values("amount", ascending=False).head() # padrão do Pandas é ordem CRESCENTE
```

```python
support.sort_values("amount", ascending=False).head()
```

```python
oppose.sort_values("amount", ascending=False).head()
```

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
# Agrupando a coluna "committee_name" (que possui valores idênticos em "x" e "y" pq Pandas atribui nomes distintos 
    # quando duas bases são mescladas com colunas idênticas), somando os valores de "amount", exibindo a série como Dataframe
        # organizando os valores em ordem decrescente 
```

```python
merged.groupby(["contributor_firstname", "contributor_lastname"]).amount.sum().reset_index().sort_values("amount", ascending=False)
   # Agrupando primeiro e último nomes dos contribuintes
```

```python
merged.groupby(["contributor_firstname", "contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
 # Descobrindo o posicionamento dos maiores doadores
```

If we want to chart out the top supporters of the proposition, we first need to select them from the dataset. Using the grouping and sorting tricks we learned earlier, the top 10 can returned like this:

```python
top_supporters = support.groupby(["contributor_firstname", "contributor_lastname"]).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
top_supporters.head(10)
```

```python
top_supporters.amount.plot.bar() # Plotando um gráfico de barras
```

```python
top_supporters.head(5).amount.plot.barh() # limitando a 5 e invertendo a orientação de exibição
```

What are those y axis labels? Those are the row number (pandas calls them indexes) of each row. We don’t want that. We want the names. We can swap them in by saving our chart to a variable and then using another matplotlib option, set_yticklabels to instruct the system which field to use.

```python
chart = top_supporters.head(5).amount.plot.barh()
chart.set_yticklabels(top_supporters.contributor_lastname)
```

```python
top_supporters['contributor_fullname'] = top_supporters.contributor_firstname + " " + top_supporters.contributor_lastname 
    # unindo nome e sobrenome 
```

```python
top_supporters.head()
```

```python
# Substituindo a etiqueta para o nome completo
chart = top_supporters.head(5).amount.plot.barh()
chart.set_yticklabels(top_supporters.contributor_fullname)
```

```python
top_supporters.head(5).to_csv("top_supporters.csv")
```

```python
# Limpeza
def combine_names(row):
    if row.contributor_fullname.startswith('SEAN PARKER'):
        return 'SEAN PARKER'
    return row.contributor_fullname
```

```python
top_supporters.apply(combine_names, axis=1)
```

```python
top_supporters['contributor_cleanname'] = top_supporters.apply(combine_names, axis=1)
```

```python
top_supporters.groupby("contributor_cleanname").amount.sum().reset_index().sort_values("amount", ascending=False).head(10)

```
